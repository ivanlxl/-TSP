```python
import csv
from tabulate import tabulate
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('AntColony.txt')
csvreader = csv.reader(file)
tourOrder = []
for row in csvreader:
    tourOrder.append(row)
tourOrder
file.close()

header = ["City", "X Coordinate", "Y Coordinate"]
print(tabulate(tourOrder, headers=header))

city = [int(row[0]) for row in tourOrder]
X = [float(row[1]) for row in tourOrder]
Y = [float(row[2]) for row in tourOrder]

plt.figure(figsize=(12, 10))
plt.plot(X, Y, marker='o', c='blue', mfc='red', mec='k')

plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Ant Colony Optimization Algorithm')

for i in range(len(X)):
    if i == 0:
        plt.text(X[i], Y[i], city[i], ha='left', va='bottom', color='blue', fontsize=12, fontweight='bold')
    elif i == len(X) - 1:
        plt.text(X[i], Y[i], city[i], ha='left', va='bottom', color='red', fontsize=12, fontweight='bold')
    else:
        plt.text(X[i], Y[i], city[i], ha='left', va='bottom')

plt.xticks(np.arange(min(X), max(X)+1, 4000))
plt.show()
```
